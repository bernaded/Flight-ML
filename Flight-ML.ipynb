## Check Outlier

```python
def find_outlier_boundary(df, variable):

    # calculate outer limit that is outlier

    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

    lower_boundary = df[variable].quantile(0.25) - (IQR * 1.5)
    upper_boundary = df[variable].quantile(0.75) + (IQR * 1.5)

    return upper_boundary, lower_boundary
```

```python
# limit for LSTAT

col_transform = ['FFP_TIER', 'AGE', 'FLIGHT_COUNT', 'BP_SUM', 'SUM_YR_1', 'SUM_YR_2', 'SEG_KM_SUM', 'LAST_TO_END',]
col_upper_limit, col_lower_limit = find_outlier_boundary(df, col_transform)
col_upper_limit, col_lower_limit
```

```python
# replace the outlier with maximum and minimum limitation

df[col_transform]= np.where(df[col_transform] > col_upper_limit, col_upper_limit,
                       np.where(df[col_transform] < col_lower_limit, col_lower_limit, df[col_transform]))
```

```python
# memvisualisasikan outlier di LSTAT

# Membuat visualisasi boxplot
plt.figure(figsize=(15, 7))

for i, col in enumerate(col_transform, 1):
  plt.subplot(2, 4, i)
  sns.boxplot(data=df, x=col)
  plt.subplots_adjust(hspace=0.5)

plt.show()
```

```python
#Making function to visualize top 20 work provice and top 5 countries
def plot(df, column_name, n=20): #default n = 20, we can customize it.
    top = df[column_name].value_counts().head(n)
    ax = top.plot(kind='barh', figsize=(8, 6))
    plt.grid(False)

    # Adding value on each bar
    for i, v in enumerate(top):
        ax.text(v, i, str(v), ha='left', va='center')
```

# RFM dengan K-Means

```python
inertia = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(knn_new_df.values)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(20, 10))

sns.lineplot(x=range(1, 11), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(1, 11), y=inertia, s=300, color='#800000',  linestyle='--')
plt.title('Elbow Method - Finding K value Clusters')
plt.xlabel('The Amount of K')
plt.ylabel('Intertia / WSS')
```

```python
# According to the elbow graph output, the amount of chosen K is 3 ,
# due to the first point before the following points has a constant gap or
# a point after the shape curve before flattening out
from sklearn.cluster import KMeans
kmeans_elbow = KMeans(n_clusters=3, random_state=0)
kmeans_elbow.fit(knn_new_df.values)
```

```python
# Put the k-means outcome using elbow method to cluster_elbow column
knn_new_df['cluster_elbow'] = kmeans_elbow.labels_
```

```python
sns.catplot(
    y = 'cluster_elbow',
    data = knn_new_df,
    kind = 'count',
    palette = "flare",
    height = 3
).ax.set_title("Total Passengers for Each Segment")
```

Cluster 0 memiliki nilai yang sangat tinggi mencapai lebih dari 30000, sedangkan cluster 2 hanya mencapai sekitar 50% dari cluster 0.

```python
sns.pairplot(knn_new_df, hue="cluster_elbow", palette="bright", height=2, aspect=1)
plt.show()
```

**Penjelasan Pairplot**
1. Frequency vs Recency: cluster 0 dan cluster 1 memiliki tingkat recency yang lebih rendang dibanding dengan cluster 2. Namun, tingkat frequency cluster 1 lebih tinggi dibandingkan dengan cluster 0 dan cluster 2.
2. Monetary vs Recency: cluster 2 memiliki rentang nilai yang lebih lebar pada nilai recency dan monetary dibanding dengan cluster 0 dan cluster 1.
3. Monetary vs Frequency: cluster 1 memiliki sebaran yang merata pada parameter frequency dan monetary, sedangkan cluster 2 hanya memiliki sedikit sebaran baik pada frequency dan monetary.
